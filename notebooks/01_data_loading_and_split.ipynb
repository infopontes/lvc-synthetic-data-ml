```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
```

```python
SEEDS = [41, 42, 46]
TEST_SIZE = 0.3

TARGET = "diagnosis"
```

```python
df = pd.read_csv("../data/raw/leish_dataset.csv")

print("Shape:", df.shape)
df.head()
```

```python
print(df[TARGET].value_counts())
print(df[TARGET].value_counts(normalize=True))
```

```python
X = df.drop(columns=[TARGET])
y = df[TARGET]
```

```python
splits = {}

for seed in SEEDS:
    X_train, X_test, y_train, y_test = train_test_split(
        X,
        y,
        test_size=TEST_SIZE,
        stratify=y,
        random_state=seed
    )

    splits[seed] = {
        "X_train": X_train,
        "X_test": X_test,
        "y_train": y_train,
        "y_test": y_test
    }

    print(f"\nSeed: {seed}")
    print("Train shape:", X_train.shape)
    print("Test shape:", X_test.shape)
    print("Train distribution:\n", y_train.value_counts(normalize=True))
    print("Test distribution:\n", y_test.value_counts(normalize=True))
```

```python
for seed in SEEDS:
    train_idx = set(splits[seed]["X_train"].index)
    test_idx = set(splits[seed]["X_test"].index)

    intersection = train_idx.intersection(test_idx)

    print(f"Seed {seed} - Overlap:", len(intersection))
```
